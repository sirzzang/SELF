## 글 제목 말머리 확인

* 두 글자 말머리 : '정치', '경제', '사회', '문화', '코인', '음악', '연예', '방송', '게임', '뻘글', '후기', '채팅', '짤방', '단문', '펌글', '질문', '응원', '키움', '두산', '롯데', '삼성', '한화', 'kt', 'LG', 'NC', 'SK'
* 세 글자 말머리 : '스포츠', '아이돌', 'LOL', '19금', '17금', 'COB', '주번나', 'KIA', 'KBO'
* 네 글자 말머리 : '개선요청', '아마야구'

```
df = pd.read_csv("/content/drive/My Drive/파이널가쥬앗/[작업 내용]/MLBPARK_주52시간_49pages.csv")
display(df)
mlbpark = df.copy()
```

```
# 말머리 확인

print(mlbpark["title"].str[:2].unique()) # 거의 다 두 글자 말머리 안에 들어간다. 일단 두 글자 말머리부터 없애자.
print(mlbpark["title"].str[:3].unique())
print(mlbpark["title"].str[:4].unique())
```

```
headers = ['정치', '경제', '사회', '문화', '코인', '음악', '연예', '방송', '게임', 
           '뻘글', '후기', '채팅', '짤방', '단문', '펌글', '질문', '응원',
           '키움', '두산', '롯데', '삼성', '한화', 'kt', 'LG', 'NC', 'SK',
           '스포츠', '아이돌', 'LOL', '19금', '17금', 'COB', '주번나', 'KIA', 'KBO', '[펌]'
           '개선요청', '아마야구']

def cleanseTitle(df, colname, lst):
    for i in range(len(df)):
        if df[colname][i][:2] in lst:
            df[colname][i] = df[colname][i][2:]
        if df[colname][i][:3] in lst:
            df[colname][i] = df[colname][i][3:]
        if df[colname][i][:4] in lst:
            df[colname][i] = df[colname][i][4:]
    return df

cleanseTitle(mlbpark, 'title', headers)
```
